```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load dataset
df = pd.read_csv("VIW_FNT.csv")
```

```python
# Group by country and year and sum the number of cases
df_cleaned = df.groupby(["WHOREGION", "COUNTRY_AREA_TERRITORY", "ISO_YEAR"], as_index=False)["INF_A"].sum()
df_cleaned = df_cleaned[df_cleaned.ISO_YEAR != 2025]
```

```python
print(df_cleaned)
```

```python
# Sum the cases for all countries for each year
df_yearly = df_cleaned.groupby("ISO_YEAR", as_index=False)["INF_A"].sum()

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df_yearly["ISO_YEAR"], df_yearly["INF_A"], marker='o')
plt.xlabel("Year")
plt.ylabel("Total Cases")
plt.title("Total Cases by Year")
plt.grid(True)
plt.show()
```
